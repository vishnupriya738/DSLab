```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11) 
y = x**2 
x
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)

plt.plot(
    x, x + 3,
    color='b',
    lw=3,
    linestyle='dashed',
    marker='o',
    markersize=12,
    markeredgecolor='green',
    markeredgewidth=4,
    markerfacecolor='red'
)

plt.plot(
    x, x + 4,
    color='r',
    lw=4,
    linestyle='dashed',
    marker='o',
    markersize=10,
    markeredgecolor='blue',
    markeredgewidth=5,
    markerfacecolor='green'
)

plt.show()
```
